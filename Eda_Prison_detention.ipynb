```python
removal_list = data_import.loc[(data_import["CAPACITY"] == -999) | (data_import["POPULATION"] == -999)].index.tolist()
```

```python
data_import["CAPACITY"].replace(-999,pd.NA,inplace=True)
data_import["POPULATION"].replace(-999,pd.NA,inplace=True)
```

```python
data_import["CAPACITY"] = data_import["CAPACITY"].fillna(data_import["CAPACITY"].median())
data_import["POPULATION"] = data_import["POPULATION"].fillna(data_import["POPULATION"].median())

```
